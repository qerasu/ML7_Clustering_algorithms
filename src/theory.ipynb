# Part 1

## 1. Инициализация центроидов в K-means

**Мой вариант:** комбинация разброса и плотности.

1. Беру случайную точку как первый центр.
2. Дальше выбираю следующий центр из точек, которые:
   - далеко от уже выбранных центров (максимизируем расстояние),
   - и находятся в зоне нормальной плотности (не выбросы).

Беру точки, которые хорошо разнесены, но не одиночные.

**Плюсы:**
- меньше риска, что центр попадет в выброс;
- центры получаются более равномерно распределенными;
- описанный алгоритм должен сходиться быстрее, чем просто случайный.

**Минусы:**
- нужно посчитать расстояния и локальную плотность, что будет чуть дороже;
- результат все равно зависит от случайного старта.

**Forgy (случайные точки как центры):**
- Плюсы: очень просто и быстро.
- Минусы: может выбрать выбросы и близкие точки, качество нестабильно.

**Random Partition (случайное распределение по кластерам, центры = средние):**
- Плюсы: тоже просто, центры являются средними величинами по кластерам.
- Минусы: сильная случайность, может дать плохие стартовые кластеры.

## 2. Как ускорить KNN с помощью кластеризации (идея kd-tree)

Можно сначала разбить данные на кластеры (например, K-means), а потом:
- при запросе искать ближайший кластер к точке;
- KNN применять только внутри этого кластера (или пары ближайших кластеров).

Так уменьшится количество точек для перебора.

Это похоже на идею **kd-tree**: алгоритм делит пространство по координатам и позволяет
быстро отсеивать далёкие области. Кластеризация сужает область поиска.
